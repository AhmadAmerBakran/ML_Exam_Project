## Goal

This notebook solves a binary classification task using the Adult Census Income dataset.

The goal is to predict whether a person earns more than 50K USD per year (`>50K`) or not (`<=50K`) based on census information.

Because this is a tabular dataset, the project examines and compares two model architectures:

1. Random Forest Classifier
2. Histogram-Based Gradient Boosting Classifier

## 1. Imports and global settings

The project uses `pandas` and `numpy` for data handling, `scikit-learn` for preprocessing, model training and evaluation, and `matplotlib` for simple visualizations.

```python
import os
import warnings
from pathlib import Path

import joblib
import numpy as np
import pandas as pd


from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    precision_score,
    recall_score,
    roc_auc_score,
    RocCurveDisplay,
)
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

warnings.filterwarnings("ignore")

RANDOM_STATE = 42
CV_FOLDS = 3
N_ITER = 15

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
RESULTS_DIR = PROJECT_ROOT / "results"
MODELS_DIR = PROJECT_ROOT / "models"
RESULTS_DIR.mkdir(exist_ok=True)
MODELS_DIR.mkdir(exist_ok=True)

print('Importing done')
```

## 2. Load the dataset

The Adult dataset has an official training file and an official test file.

- `adult.data` is used for model training and cross-validation.
- `adult.test` is used only for final evaluation.

```python
COLUMN_NAMES = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education_num",
    "marital_status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital_gain",
    "capital_loss",
    "hours_per_week",
    "native_country",
    "income",
]

TRAIN_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
TEST_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test"

train_df = pd.read_csv(
    TRAIN_URL,
    names=COLUMN_NAMES,
    na_values="?",
    skipinitialspace=True,
)

test_df = pd.read_csv(
    TEST_URL,
    names=COLUMN_NAMES,
    na_values="?",
    skipinitialspace=True,
    skiprows=1,
)

print("Training shape:", train_df.shape)
print("Test shape:", test_df.shape)
train_df.head()
```

## 3. Data preparation

The dataset is tabular, so the data preparation must be more than simply loading the data.

The preparation below includes:

1. Cleaning the target labels.
2. Removing duplicate rows from the training data.
3. Creating new features from capital gain and capital loss.
4. Dropping columns that are less useful or redundant.
5. Separating features and labels.
6. Building preprocessing pipelines for numeric and categorical features.

```python
def clean_income_label(df: pd.DataFrame) -> pd.DataFrame:
    """Clean the target column and convert it to a binary label."""
    df = df.copy()
    df["income"] = df["income"].astype(str).str.strip().str.replace(".", "", regex=False)
    df["income"] = df["income"].map({"<=50K": 0, ">50K": 1})
    return df


def add_engineered_features(df: pd.DataFrame) -> pd.DataFrame:
    """Add simple, explainable features from capital gain/loss."""
    df = df.copy()
    df["capital_net"] = df["capital_gain"] - df["capital_loss"]
    df["has_capital_gain"] = (df["capital_gain"] > 0).astype(int)
    df["has_capital_loss"] = (df["capital_loss"] > 0).astype(int)
    return df


train_df = clean_income_label(train_df)
test_df = clean_income_label(test_df)

before_duplicates = len(train_df)
train_df = train_df.drop_duplicates().reset_index(drop=True)
after_duplicates = len(train_df)

train_df = add_engineered_features(train_df)
test_df = add_engineered_features(test_df)

print(f"Removed duplicate training rows: {before_duplicates - after_duplicates}")
print("Missing values in training data:")
display(train_df.isna().sum().sort_values(ascending=False).head(10))
```

```python
# We drop education because education_num already contains the ordered education level.
# We drop fnlwgt because it is a census sampling weight and is less suitable as a normal predictive feature.
DROP_COLUMNS = ["education", "fnlwgt"]
TARGET_COLUMN = "income"

X_train = train_df.drop(columns=DROP_COLUMNS + [TARGET_COLUMN])
y_train = train_df[TARGET_COLUMN]

X_test = test_df.drop(columns=DROP_COLUMNS + [TARGET_COLUMN])
y_test = test_df[TARGET_COLUMN]

numeric_features = X_train.select_dtypes(include=["int64", "float64", "int32", "float32"]).columns.tolist()
categorical_features = X_train.select_dtypes(include=["object", "category"]).columns.tolist()

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)
print("Training target distribution:")
print(y_train.value_counts(normalize=True).rename({0: "<=50K", 1: ">50K"}))
```

## 4. Preprocessing pipeline

The preprocessing is handled by `ColumnTransformer`.

Numeric columns:

- missing values are replaced with the median value.

Categorical columns:

- missing values are replaced with the most frequent category.
- categories are converted to numbers using one-hot encoding.

One-hot encoding is needed because the models cannot directly understand text categories such as `Private`, `Married-civ-spouse`, or `Exec-managerial`.

```python
def make_one_hot_encoder():
    """Create a OneHotEncoder that works across common scikit-learn versions."""
    try:
        return OneHotEncoder(handle_unknown="ignore", min_frequency=10, sparse_output=False)
    except TypeError:
        try:
            return OneHotEncoder(handle_unknown="ignore", min_frequency=10, sparse=False)
        except TypeError:
            return OneHotEncoder(handle_unknown="ignore", sparse=False)


numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", make_one_hot_encoder()),
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)
```

## 5. Model 1 - Random Forest

A Random Forest is an ensemble model made of many decision trees.

Each tree is trained slightly differently, and the forest combines their predictions. This usually gives better generalization than one single decision tree.

The model is fine-tuned using `RandomizedSearchCV` with 3-fold cross-validation.

```python
rf_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        (
            "classifier",
            RandomForestClassifier(
                random_state=RANDOM_STATE,
                n_jobs=-1,
            ),
        ),
    ]
)

rf_param_distributions = {
    "classifier__n_estimators": [200, 400, 600],
    "classifier__max_depth": [None, 10, 20, 30, 40],
    "classifier__min_samples_split": [2, 5, 10],
    "classifier__min_samples_leaf": [1, 2, 4],
    "classifier__max_features": ["sqrt", "log2", None],
    "classifier__class_weight": [None, "balanced"],
}

rf_search = RandomizedSearchCV(
    estimator=rf_pipeline,
    param_distributions=rf_param_distributions,
    n_iter=N_ITER,
    scoring="f1",
    cv=CV_FOLDS,
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbose=1,
)

rf_search.fit(X_train, y_train)

print("Best Random Forest CV F1:", rf_search.best_score_)
print("Best Random Forest parameters:")
print(rf_search.best_params_)
```

## 6. Model 2 - Histogram-Based Gradient Boosting

Gradient Boosting also uses decision trees, but in a different way than Random Forest.

A Gradient Boosting model builds trees sequentially. Each new tree tries to correct errors made by the previous trees.

Histogram-Based Gradient Boosting is a faster version that groups continuous values into bins. It often performs very well on structured/tabular datasets.

This model is also fine-tuned using `RandomizedSearchCV`.

```python
hgb_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        (
            "classifier",
            HistGradientBoostingClassifier(
                random_state=RANDOM_STATE,
            ),
        ),
    ]
)

hgb_param_distributions = {
    "classifier__learning_rate": [0.03, 0.05, 0.1, 0.15],
    "classifier__max_iter": [100, 200, 300],
    "classifier__max_leaf_nodes": [15, 31, 63],
    "classifier__min_samples_leaf": [10, 20, 50],
    "classifier__l2_regularization": [0.0, 0.1, 1.0],
}

hgb_search = RandomizedSearchCV(
    estimator=hgb_pipeline,
    param_distributions=hgb_param_distributions,
    n_iter=N_ITER,
    scoring="f1",
    cv=CV_FOLDS,
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbose=1,
)

hgb_search.fit(X_train, y_train)

print("Best Histogram-Based Gradient Boosting CV F1:", hgb_search.best_score_)
print("Best Histogram-Based Gradient Boosting parameters:")
print(hgb_search.best_params_)
```

## 7. Final evaluation on the official test set

The official test set has not been used during training or fine-tuning.

This final evaluation gives a fairer estimate of how the models perform on unseen data.

The main metrics are:

- **Accuracy:** how often the model is correct overall.
- **Precision:** out of the people predicted as `>50K`, how many actually are `>50K`.
- **Recall:** out of the people who actually are `>50K`, how many the model found.
- **F1-score:** balances precision and recall.
- **ROC-AUC:** measures how well the model separates the two classes across thresholds.

```python
def evaluate_model(model_name: str, model, X, y) -> dict:
    """Evaluate a fitted model and return the main metrics."""
    y_pred = model.predict(X)
    y_proba = model.predict_proba(X)[:, 1]

    metrics = {
        "model": model_name,
        "accuracy": accuracy_score(y, y_pred),
        "precision": precision_score(y, y_pred),
        "recall": recall_score(y, y_pred),
        "f1": f1_score(y, y_pred),
        "roc_auc": roc_auc_score(y, y_proba),
    }

    print()
    print(f"{model_name}")
    print("-" * len(model_name))
    print(classification_report(y, y_pred, target_names=["<=50K", ">50K"]))

    return metrics


best_rf = rf_search.best_estimator_
best_hgb = hgb_search.best_estimator_

results = [
    evaluate_model("Random Forest", best_rf, X_test, y_test),
    evaluate_model("Histogram-Based Gradient Boosting", best_hgb, X_test, y_test),
]

results_df = pd.DataFrame(results).sort_values(by="f1", ascending=False).reset_index(drop=True)
results_df
```

```python
results_path = RESULTS_DIR / "model_comparison.csv"
results_df.to_csv(results_path, index=False)
print(f"Saved model comparison to: {results_path}")
```
